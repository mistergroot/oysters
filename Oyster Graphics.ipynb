```python
import csv
from wordcloud import WordCloud
import matplotlib.pyplot as plt

reader = csv.reader(open('/home/nickpichome/Desktop/Untitled1.csv', 'r',newline='\n'))
d = {}
for k,v in reader:
    d[k] = int(v)

#Generating wordcloud. Relative scaling value is to adjust the importance of a frequency word.
#See documentation: https://github.com/amueller/word_cloud/blob/master/wordcloud/wordcloud.py
wordcloud = WordCloud(width=2000,height=1080, max_words=1628, colormap="GnBu", background_color="black",relative_scaling=1,normalize_plurals=False).generate_from_frequencies(d).to_file("/home/nickpichome/Desktop/index.png")

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
